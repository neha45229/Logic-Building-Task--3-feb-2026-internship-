1.Employee Performance Bonus Eligibility

```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
top_score = max(employees.values())
top_performers=[]
for name in employees:
  if employees[name] == top_score:
    top_performers.append(name)
print("Top Performers Eligible for Bonus:",",".join(top_performers),"(Score:",top_score,")")
```

2.Search Query Keyword Analysis

```python
sentence=input().lower()
words=sentence.split()
result = { }
for w in words:
  if words.count(w)>1:
    result[w]=words.count(w)
print(result)





```

3.Sensor Data Validation


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings =[]
for i in range(len(sensor_readings)):
  if sensor_readings[i] % 2 == 0:
    valid_readings.append((i,sensor_readings[i]))
print("Valid Sensor Readings (Hour,Value):")
print(valid_readings)
```

4.Email Domain Usage Analysis


```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
count=0
for email in emails:
  if email.split("@")[1]=="gmail.com":
    count+=1
print(f"gmail: {int((count/len(emails))*100)}%")
print(f"yahoo: {int(((len(emails)-count)/len(emails))*100)}%")

```

5. Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
avg=sum(sales)/len(sales)
for i in range(len(sales)):
  if sales[i]>avg*1.10:
    print("Day",i+1,":",sales[i])
```
